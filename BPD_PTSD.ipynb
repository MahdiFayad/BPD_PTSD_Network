# Only subjects with complete observations (n = 33287)

```python
# Filter : remove people with unknown values
# Custom function to check if '9' or 9 (as an integer) is present in a cell
def contains_9(cell):
    return ('9' in str(cell)) or (cell == 9)
df_article = df[cibles]
condition = ~df_article.applymap(contains_9).any(axis=1)
filtered_df = df_article[condition]
filtered_df
```

# Impute 0 for subjects who were not asked certains PTSD questions because they didnt meet the primary criterion 

```python
for col in filtered_df.columns : 
    filtered_df[col]=filtered_df[col].replace(['2'],0)
    filtered_df[col]=filtered_df[col].replace(['1'],1)
    filtered_df[col]=filtered_df[col].replace([' '],0)
    
for col in filtered_df.columns : 
    filtered_df[col] = filtered_df[col].astype(str).astype(int)
```

# Regroup questions based on DSM-IV symptoms

```python
def categorise(row,col,col2):  
    if row[col] > 0 or row[col2] > 0:
        return 1
    else :
        return 0

def categorise4(row,col,col2,col3,col4):  
    if row[col] > 0 or row[col2] > 0 or row[col3] > 0 or row[col4]>0:
        return 1
    else :
        return 0

def categorise3(row,col,col2,col3):  
    if row[col] > 0 or row[col2] > 0 or row[col3] > 0:
        return 1
    else :
        return 0
    
def categorise9(row,col,col2,col3,col4,col5,col6,col7,col8,col9):  
    if row[col] > 0 or row[col2] > 0 or row[col3] > 0 or row[col4] > 0 or row[col5] > 0 or row[col6] > 0 or row[col7] > 0 or row[col8] > 0 or row[col9] > 0:
        return 1
    else :
        return 0
```

```python
def change_df_ptsd(df):
    df['TRAUMA'] = df['w2s12q7b']
    df['FEAR'] = df['w2s12q7a']
    df['MEMORIES'] = df['w2s12q8a1']
    df['NIGHTMARES'] = df['w2s12q8a2']
    df['FLASHBACKS'] = df.apply(lambda row: categorise(row,'w2s12q8a3','w2s12q8a4'), axis=1)
    df['PSYCHDISTRESS'] = df['w2s12q8a5']
    df['STIMREACTIVITY'] = df['w2s12q8a6']
    df['INTERNALAVOID'] = df.apply(lambda row: categorise(row,'w2s12q8a7','w2s12q8a8'), axis=1)
    df['EXTERNALAVOID'] = df['w2s12q8a9']
    df['AMNESIA'] = df['w2s12q8a10']
    df['INTERESTREDUCT'] = df['w2s12q8a11']
    df['DETACHMENT'] = df['w2s12q8a12']
    df['AFFECTRESTRIC'] = df['w2s12q8a13']
    df['FUTURE'] = df['w2s12q8a14']
    df['DIFFSLEEPING'] = df['w2s12q8a15']
    df['IRRITANGER'] = df['w2s12q8a16']
    df['DIFFCONCENT'] = df['w2s12q8a17']
    df['HYPERVIGILANCE'] = df['w2s12q8a18']
    df['STARTLE'] = df['w2s12q8a19']
    return df
```

```python
def change_df_border(df):
    df['ABANDONMENT'] = df.apply(lambda row: categorise(row,'w2s10q1a22','w2s10q1a32'), axis=1)
    df['INTERPERSONALINSTAB'] = df['w2s10q1a23']
    df['IDENTITY'] = df.apply(lambda row: categorise4(row,'w2s10q1a24','w2s10q1a25','w2s10q1a26','w2s10q1a27'), axis=1)
    df['IMPULSIVITY'] = df.apply(lambda row: categorise3(row,'w2s10q1a20','w2s10q1a21','w2s10q1a28'), axis=1)
    df['AUTOAGRESSIVITY'] = df.apply(lambda row: categorise(row,'w2s10q1a29','w2s10q1a30'), axis=1)
    df['MOODSWING'] = df['w2s10q1a31']
    df['EMPTY'] = df['w2s10q1a33']
    df['ANGER'] = df.apply(lambda row: categorise3(row,'w2s10q1a34','w2s10q1a35','w2s10q1a36'), axis=1)
    df['PARANOID'] = df['w2s10q1a37']
    return df
```

```python
concat_intitulés_border = ['MOODSWING', 'EMPTY','IDENTITY','ABANDONMENT', 'INTERPERSONALINSTAB','IMPULSIVITY','ANGER', 'PARANOID', 'AUTOAGRESSIVITY']
concat_intitulés_ptsd = ['TRAUMA','FEAR','MEMORIES','NIGHTMARES','FLASHBACKS','PSYCHDISTRESS','STIMREACTIVITY','INTERNALAVOID',
                         'EXTERNALAVOID','AMNESIA',
                        'INTERESTREDUCT',
                         'DETACHMENT',
                         'AFFECTRESTRIC','FUTURE','DIFFSLEEPING','IRRITANGER','DIFFCONCENT','HYPERVIGILANCE','STARTLE']

concat_final = concat_intitulés_border+concat_intitulés_ptsd
```

```python
filtered_df = change_df_border(filtered_df)
filtered_df = change_df_ptsd(filtered_df)

final_df = filtered_df[concat_final]
```
